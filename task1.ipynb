```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population = pd.read_csv("world_population.csv")
population.head()
```

"DATA EXPLORATION"

```python
# (1) Getting the dataset shape:
print(population.shape)
```

```python
# (2) Checking if there's any NULL values in the dataset:
population.isnull().values.any()
```

```python
# (3) Getting the data infos:
print(population.info())
```

```python
# (4) Sorting the dataset descendingly according to its "Growth Rate":
top_gr = population.sort_values(by="Growth Rate", ascending=False)
top_gr.head()
```

```python
# (5) Sorting the dataset descendingly according to its "2022 population":
top_2022pop = population.sort_values(by="2022 Population", ascending=False)
top_2022pop.head()
```

```python
population["1970 Population"].mean()
```

- Top 5 countries at 2022 Population:

- Average population per country in 2022 = 34,074,414
- Average population per country in 1970 = 15,786,908


```python
# Define the width of each bar
bar_width = 0.25

# Set the position of each bar on the x-axis
x = np.arange(len(top_2022pop))

# Plotting the bars
plt.figure(figsize=(10, 10))
plt.bar(x - bar_width/2, top_2022pop["2022 Population"], width=bar_width, label='2022 Population')
plt.bar(x + bar_width/2, top_2022pop["1970 Population"], width=bar_width, label='1970 Population')

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Population in Billions')
plt.title('Population in 2022 VS 1970')
plt.xticks(x, top_2022pop['Country/Territory'])
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

India looks like it has a HUGE growth rate, let's see its population over the years:

```python
years = list(population.columns[5:12])
years = [year.replace(" Population",'') for year in years]

population = list(population[population['Country/Territory'] == "India"].iloc[:,5:12].values.flatten())

plt.figure(figsize=(10, 6))
plt.bar(years, population, color='skyblue',alpha=.9)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of India over the years')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.style.use('dark_background')
plt.show()
```
